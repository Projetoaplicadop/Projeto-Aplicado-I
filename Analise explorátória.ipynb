```python
import pandas as pd

try:
  df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/idebrmesaopaulo.xlsx')

  # Remove as linhas 0, 1 e 2
  df = df.iloc[3:]

  # Use o método 'style' para melhorar a visualização
  styled_df = df.style.set_table_styles([
      {'selector': 'th', 'props': [('background-color', '#f0f0f0'), ('color', 'black'), ('font-weight', 'bold')]},  # Estiliza o cabeçalho
      {'selector': 'td', 'props': [('text-align', 'center')]}, # centraliza o texto
  ])

  display(styled_df)
except FileNotFoundError:
  print("Arquivo não encontrado. Verifique o nome do arquivo e o caminho.")
except Exception as e:
  print(f"Ocorreu um erro: {e}")
```
